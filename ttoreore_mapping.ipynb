```python
url = 'https://www.toreore.com/controller/ajax_find_area_controller.html'
data={
    'actionMethod':'get_sido'
}

response = requests.post(url,data=data)
data = json.loads(response.text)

sigugun_list = []
listdata = data['Listdata']
for i in range(0,len(listdata)):
    sidoid = listdata[i]['sidoid']
    sidoname = listdata[i]['name']
    
    url = 'https://www.toreore.com/controller/ajax_find_area_controller.html'
    data={
        'actionMethod': 'get_sigugun',
        'cur_sido': sidoid
    }
    response = requests.post(url,data=data)
    jsondatas = json.loads(response.text)
    
    jsondata = jsondatas['Listdata']
    for i in range(0,len(jsondata)):
        name = jsondata[i]['name']
        sigugun_list.append([sidoname,sidoid,name])
print(sigugun_list)
```

```python
lists = []
for i in range(0,len(sigugun_list)):
    url = 'https://www.toreore.com/controller/ajax_shop_search_controller.html'
    data={
        'sido': sigugun_list[i][0],
        'sigugun':sigugun_list[i][2],
        'actionMethod':'search',
        'dataindex':0
    }
    
    response = requests.post(url,data=data)
    jsondatas = json.loads(response.text)
    jsondata = jsondatas['Listdata']
    
    for i in range(0,len(jsondata)):
        addr = jsondata[i]['addr1']
        name = jsondata[i]['sname']
        phone = jsondata[i]['phone1']
        lat = jsondata[i]['lat']
        lng = jsondata[i]['lng']
        lists.append([addr,name,phone,lat,lng])
lists
```

```python
lati_lists = []
for list in lists:
    url = 'https://dapi.kakao.com/v2/local/search/address.json'
    kakao_key = '3664387079752a9ac90a4f3189f5ba07'
    headers = {'Authorization': 'KakaoAK ' + kakao_key}
    address = list[0]
    params = {'query': address}

    response = requests.get(url, headers=headers, params=params)

    if response.status_code == 200:
        data = json.loads(response.text)
    
    documents = data.get('documents',[])
    
    if documents:
        docu = documents[0]
        x = docu['x']
        y = docu['y']
        lati_lists.append([x,y])
    else:
        pass
    
lati_lists
```
